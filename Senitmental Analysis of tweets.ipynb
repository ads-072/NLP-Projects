```python
import pandas as pd
```

```python
# read data csv file 
data = pd.read_csv('Twitter_Tweets.csv')
data
```

```python
data.shape
```

```python
# drop duplicate values
data.drop_duplicates(inplace = True)
```

```python
#Code to remove @
data['tweet'] = data['tweet'].apply(lambda x : ' '.join([tweet for tweet in x.split()if not tweet.startswith("@")]))
data
```

```python
# remove numerical data
data['tweet'] = data['tweet'].apply(lambda x : ' '.join([tweet for tweet in x.split() if not tweet == '\d*']))
data
```

```python
#pip install unidecode
import unidecode
```

```python
#Removing all the greek characters using unidecode library
data['tweet'] = data['tweet'].apply(lambda x : ' '.join([unidecode.unidecode(word) for word in x.split()])) 
data
```

```python
#Removing the word 'hmm' and it's variants
data['tweet'] = data['tweet'].apply(lambda x : ' '.join([word for word in x.split() if not word == 'h(m)+' ]))
data
```

```python
# remove stopwords
from nltk.corpus import stopwords
data['tweet'] = data['tweet'].apply(lambda x : ' '.join([word for word in x.split() if not word in set(stopwords.words('english'))]))
data
```
